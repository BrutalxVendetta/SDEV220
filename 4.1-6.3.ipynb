```python
#6.2
# The code will repeatedly compare 2 numbers, it increases the "Number" until it matches the "Guess"

guess = 7
number = 1

while True:
    if number < guess:
        print("Too Low")
    elif number == guess:
        print("Correct")
        break
    else:
        print("Oops")
        break
    number += 1
```
